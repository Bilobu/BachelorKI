```python
import nltk 
from nltk.corpus import stopwords
# um spezielle Zeichen wie '.' oder '--' zu elimineren werden diese zur stopword list von nltk im root verzeichnis hunzugefügt
specialChars = ['--',':',',','>','<',"''",'.','´´','``','+','!']
stopWords = nltk.corpus.stopwords.words('german')
for char in specialChars:
    stopWords.append(char)

print (stopWords)

#stopwords entfernen
def filter(wordsToFilter):
    
    filtered = [w for w in wordsToFilter if not w in stopWords]
    #"".join(filtered)
    #print (filtered)
    return filtered
    
    










```

```python
import nltk
import string
import os

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.porter import PorterStemmer


token_dict = {}
dic ={}
stemmer = PorterStemmer()

def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

def tokenize(text):
    tokens = nltk.word_tokenize(text)
    stems = stem_tokens(tokens, stemmer)
    return stems


for file in docs:
    lowers = file.lower()
    #test = filter(lowers)
    #print(test)
    no_punctuation = lowers.translate( string.punctuation)
    no_punctuationFilter = filter(no_punctuation)
    token_dict[file] = no_punctuation

#print (token_dict.values())

for String in docs[0]:
    lowers = String.lower()
    dic[String] = lowers

    
#print(docs[0])
    
tfidf = TfidfVectorizer(tokenizer=tokenize, stop_words = stopWords)
tfs = tfidf.fit_transform(token_dict.values())

idf = tfidf.idf_
print (dict(zip(tfidf.get_feature_names(), idf)))
```
